```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df_cereal = pd.read_csv('cereal.csv')
df_circle = pd.read_csv('circle_data.csv')
```

```python
print("--- 1. Tampilkan 20 data pertama ---")
print(df_cereal.head(20))

print("\n--- 2. Tampilkan 10 data terakhir ---")
print(df_cereal.tail(10))

print("\n--- 3. Nama kolom pada dataset ---")
print(df_cereal.columns.tolist())

print("\n--- 4. Ambillah 10 data (baris) dari kolom vitamin ---")
# Mengambil kolom 'vitamins' lalu ambil 10 teratas
print(df_cereal['vitamins'].head(10))
```

```python
print("\n--- ASSIGNMENT 1 ---")
subset_data = df_cereal[['name', 'sugars', 'weight']].iloc[3:10]
print(subset_data)
```

```python
print("\n--- ASSIGNMENT 2 ---")

# 1. Berapakah kalori dari merk Strawberry Fruit Wheats?
filter_1 = df_cereal[df_cereal['name'] == 'Strawberry Fruit Wheats']
print("Kalori Strawberry Fruit Wheats:", filter_1['calories'].values[0])

# 2. Berapakah gula dan vitamin dari merk Smacks?
filter_2 = df_cereal[df_cereal['name'] == 'Smacks']
print("Gula Smacks:", filter_2['sugars'].values[0])
print("Vitamin Smacks:", filter_2['vitamins'].values[0])

# 3. Jumlah baris dan cek data null
jumlah_baris = df_cereal.shape[0]
cek_null = df_cereal.isnull().any().any()

print(f"Jumlah baris data: {jumlah_baris}")
print(f"Adakah data yang null? {'Ya' if cek_null else 'Tidak'}")
```

```python
# Visualisasi Scatterplot
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df_circle, x='x', y='y')
plt.title("Scatterplot Circle Data dari Kaggle")
plt.show()
```

```python
#Visualisasi Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data=df_circle, x='x', kde=True)
plt.title("Histogram Circle Data dari Kaggle")
plt.show()
```

```python
print("\n--- ASSIGNMENT 3 ---")
# Filter dulu datanya
bran_data = df_cereal[df_cereal['name'] == '100% Bran']

plt.figure(figsize=(6,6))
# Plotting
sns.scatterplot(data=bran_data, x='calories', y='protein', s=200, color='red')

plt.title("Scatterplot: 100% Bran")
plt.xlabel("Kalori")
plt.ylabel("Protein")
plt.grid(True)
plt.show()
```
